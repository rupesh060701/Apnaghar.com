```python


# Concatenating relevant columns into the 'Tags' column
apartment['Tags'] = \
    apartment['PropertySubName'].astype(str) + ' | ' + \
    apartment['NearbyLocations'].astype(str) + ' | ' + \
    apartment['LocationAdvantages'].astype(str) + ' | ' + \
    apartment['PriceDetails'].astype(str) + ' | ' + \
    apartment['TopFacilities'].astype(str)

# Display the DataFrame with the new 'Tags' column
print(apartment.head())


```

```python
new_data=apartment.drop(columns=['PropertySubName','NearbyLocations','LocationAdvantages','Link','PriceDetails','TopFacilities'])
```

```python
new_data.iloc[1].Tags
```

```python
import re

def extract_list(s):
    return re.findall(r"'(.*?)'", s)

apartment['Tags'] = apartment['Tags'].apply(extract_list)
```

```python
new_data.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
cv



# from sklearn.feature_extraction.text import TfidfVectorizer
# cv = TfidfVectorizer(stop_words='english', ngram_range=(1, 2))
# cv
```

```python
vector=cv.fit_transform(new_data['Tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['PropertyName']=='M3M Golf Hills'].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].PropertyName)
```

```python
def recomendation(home):
  index=new_data[new_data['PropertyName']==home].index[0]
  if index:
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
      print(new_data.iloc[i[0]].PropertyName)
  else:
    print('Property not found')
```

```python
recomendation('M3M Golf Hills')
```

```python
def recommend_properties(property_name, cosine_sim=similarity):

    idx = apartment.index[apartment['PropertyName'] == property_name].tolist()[0]


    sim_scores = list(enumerate(similarity[idx]))


    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    sim_scores = sim_scores[1:11]


    property_indices = [i[0] for i in sim_scores]

    recommendations_df = pd.DataFrame({
        'PropertyName': apartment['PropertyName'].iloc[property_indices],
        'SimilarityScore': sim_scores
    })


    return recommendations_df
```

```python
recommend_properties("DLF The Arbour")
```
